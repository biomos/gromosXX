```python
import matplotlib.pyplot as plt
import numpy as np
def plot_sampled_states(filename):
    data = np.loadtxt(filename)
    x = data[:,0]
    y = data[:,1]
    y2= data[:,2]
    y3=[i*5+j for i,j in zip(y,y2)]
    plt.scatter(x,y3,1)
    plt.show()
    return
```
